```python
import numpy as np
from scipy.cluster.hierarchy import fcluster
from scipy.spatial.distance import euclidean
from sklearn.metrics import accuracy_score
 
with open("2b_avg_pd_equilibrium", 'rb') as pickle_file:
    avg_pd_2b = pickle.load(pickle_file)
 
with open("3b_avg_pd_equilibrium", 'rb') as pickle_file:
    avg_pd_3b = pickle.load(pickle_file)
 
print(np.shape(avg_pd_2b))
 
# Combine the arrays along the second axis (axis=1)
combined_array = np.concatenate((avg_pd_2b, avg_pd_3b), axis=1)
 
# Verify the shape of the combined array
print(combined_array.shape)  # Output will be (12, 180, 1)
 
# Reshape the array to (12, 60) before creating the DataFrame
reshaped_data = np.array(combined_array).reshape(12, 120)
 
# Create a DataFrame
df = pd.DataFrame(reshaped_data)
 
# Hierarchical clustering
linkage_matrix = linkage(df, method='median')  # You can choose a different linkage method
 
# Plot the dendrogram with labeled points
plt.figure(figsize=(10, 6))
dendrogram(linkage_matrix, labels=df.index, orientation='top', distance_sort='descending', leaf_label_func=lambda x: replace_labels(x, sp_labels))
plt.title('Dendrogram: 2B+3B P(d) Avgs.')
plt.xticks(rotation='vertical')
plt.ylabel('Distance')
plt.show()
 
# Step 1: Choose a threshold distance to form clusters
threshold_distance = 0.005  # Adjust this based on your data and clustering objectives
# Step 2: Compute cluster labels using the threshold distance
cluster_labels = fcluster(linkage_matrix, threshold_distance, criterion='distance')
print(cluster_labels) # Confirm there are 12 unique clusters
```
